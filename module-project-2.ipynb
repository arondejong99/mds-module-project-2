# Udemy business courses
In this file, a business case will be presented along with a problem the company is currently facing. Our job is to use data visualization to solve that problem. The visualizations will be created for two different groups within the company: 'CEO' and 'Head of Data Science team'.

## Business case
Udemy is a course facilitator company that offers both free and paid courses for students to upgrade their skills.

### Problem
The CEO has noticed that the amount of courses published has been falling over the last few years. The CEO is worried that this fall will influence the ratings and sales. Your job is to:
- Create insight into the rating, subscribers and sales of Udemy in order to get a better overview of the situation.
- Create visualisations that can help the head of data science to push ideas in order to improve the sales.

```python
# Importing all the modules
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Defining some important functions.    
# Create a frequency table to use for the line graphs 
def freq_table_avg(dataset, keyset, valueset):
    list_per_year = [[keyset[x], valueset[x]] for x in range(len(dataset))]
    freq_table = {x[0]: np.round(np.mean([y[1] for y in list_per_year if x[0] == y[0]]),2) 
                     for x in list_per_year}
    final_freq = {x[0]: x[1] for x in sorted(freq_table.items())}
    return final_freq

# Create function to draw line plot
def line_plot(x, y, color, marker, label, linewidth, markersize, xlabel,
             ylabel, title, xticks, gridcolor):
    plt.plot(x, y, color = color, marker = marker, label = label,
            markersize = markersize, linewidth = linewidth)
    plt.grid(color=gridcolor, linestyle='-', linewidth=2, axis='y', alpha=0.4)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.xticks(xticks)
    plt.title(title)
    plt.legend()
    plt.show()
    
# Create function to draw a scatter plot
def scatter_plot(x, y, color, alpha, title, xlabel, ylabel, label, xticks = None):
    plt.scatter(x, y, alpha = alpha, color = color, label = label)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    # Some visualizations need to have xticks
    if xticks != None:
        plt.xticks(xticks)
    plt.legend(loc = 0)
    plt.show()
```

```python
# Importing the dataset
# link to dataset: https://www.kaggle.com/jilkothari/business-courses-udemy-10k-courses
data = pd.read_csv('udemy_output_All_Business_p1_p626.csv', index_col = 0).dropna()
data.head()

# Create extra columns for the visualizations
# total_sales_eur: the total amount of sales generated by the course in euros
# the 'price_detail__amount' is in the currency of Indean Rupees. 1 EUR = 84.5 INR
conv = 84.5 
data['total_sales_eur'] = data['price_detail__amount'] * data['num_subscribers'] / conv

# discount_per: percentage of the price the discount of a course has
data['discount_per'] = data['discount_price__amount'] / data['price_detail__amount'] * 100
data.head()

# published_year: the year substracted from the 'published_time' column
data['published_year'] = [int(x[0:4]) for x in data['published_time']]

# final_prize: original price - discount in euros
data['final_prize_amount_eur'] = (data['price_detail__amount'] - data['discount_price__amount']) / conv

# remove columns that are not necessary
columns = ['title', 'url', 'is_paid', 'is_wishlisted', 'num_published_lectures', 'discount_price__price_string',
          'price_detail__price_string', 'created', 'num_published_practice_tests']
data = data.drop(columns, axis = 1)

# display the final dataset
data.head()
```

## Visualizations CEO
In this chapter, the visualizations for the CEO will be created. Because the CEO isn't as technical as the head of the data science team, the visualizations will mostly be relatively simple.

The visualizations will be as follows:
- The amount of courses published per year (`published_year`)
- Average rating of courses per year (`avg_rating`, `published_year`)
- The amount of subscribed students per year on average (`num_subscribers`, `published_year`)
- Revenue sold from courses on average per year (`num_subscribers`, `total_sales_eur`, `published_year`)

### The amount of courses published per year

```python
# Subtracting the data
published_year = [int(x) for x in data['published_year']]
years = sorted(np.unique(published_year))
freq_published_time = {int(x): published_year.count(x) for x in published_year}

final_freq_courses = {x[0]: x[1] for x in sorted(freq_published_time.items())}
print(final_freq_courses)
```

```python
# Set the parameters
x = final_freq_courses.keys()
y = final_freq_courses.values()
color = 'black'
marker = 'o'
label = "Amount of courses"
linewidth = 2.0
markersize = 9.0
gridcolor = "grey"
xlabel = "Year of publishing"
ylabel = "Amount of courses"
title = "The amount of courses \n published per year"

# Draw the line graph
line_plot(x, y, color = color, marker = marker, label = label,
          linewidth = linewidth, markersize = markersize, xlabel = xlabel,
             ylabel = ylabel, title = title, xticks = years, gridcolor = gridcolor)
```

### Findings
The amount from courses published per year has been slowly but steady increasing in Udemy's beginning years. In 2018, the amount of courses published was the highest with more than 1750 courses published. Since then, the number has been decreasing.

### Average rating of courses per year

```python
# Gather the data
ratings = np.round(np.array(data['avg_rating']), 2)
published_year = [int(x) for x in data['published_year']]
years = sorted(np.unique(published_year))

# Create the freq table
avg_per_year_freq = freq_table_avg(data, published_year, ratings)
final_avg_rating_freq = {x[0]: x[1] for x in sorted(avg_per_year_freq.items())}
print(final_avg_rating_freq)
```

```python
# Set the parameters
x = final_avg_rating_freq.keys()
y = final_avg_rating_freq.values()
color = 'green'
marker = 's'
label = "Average rating"
linewidth = 2.0
markersize = 9.0
gridcolor = "red"
xlabel = "Year"
ylabel = "Average rating"
title = "Average rating of courses per year"

# Draw the line graph
line_plot(x, y, color = color, marker = marker, label = label,
          linewidth = linewidth, markersize = markersize, xlabel = xlabel,
             ylabel = ylabel, title = title, xticks = years, gridcolor = gridcolor)
```

### Conclusion
The average rating of courses have been relatively stable since it's beginning, varieng between 4.225 and 4.0. The year 2020 has the highest average with 4.27.

### Amount of students subscribing to courses per year

Use the freq_published_time for the avg rating per year

```python
# Subtracting the data
num_subscribers = np.array(data['num_subscribers'])
published_year = [int(x) for x in data['published_year']]
years = sorted(np.unique(published_year))

# Create the freq table
freq_subscriber_time = freq_table_avg(data, published_year, num_subscribers)
final_freq_subs = {x[0]: x[1] for x in sorted(freq_subscriber_time.items())}

print((final_freq_subs))
```

```python
# Set the parameters
x = final_freq_subs.keys()
y = final_freq_subs.values()
color = 'blue'
marker = 'D'
label = "Subscribers"
linewidth = 2.5
markersize = 8.0
gridcolor = "black"
xlabel = "Amount of subscribers per course"
ylabel = "Amount of subscribers"
title = "The average amount of students \n subscribing to a course over the years"

# Draw the line graph
line_plot(x, y, color = color, marker = marker, label = label,
          linewidth = linewidth, markersize = markersize, xlabel = xlabel,
             ylabel = ylabel, title = title, xticks = years, gridcolor = gridcolor)
```

### Findings
The amount of subscribers to a course at Udemy has been plummiting from 2015 to 2019. There has been a small increase in 2020, but it's average is still not very much higher since it's beginning.

### Amount of revenue per course over the years

```python
# Get the data
sales_in_eur = np.array(data['total_sales_eur'])
published_year = [int(x) for x in data['published_year']]
years = sorted(np.unique(published_year))

# Create the freq table
sales_year_freq = freq_table_avg(data, published_year, sales_in_eur)

# Creat the final freq table. Devide each value by 1000 to make it more readable in the yticks
final_sales_freq = {x[0]: np.round(x[1] / 1000, 2) for x in sorted(sales_year_freq.items())}
print(final_sales_freq)
```

```python
# Set the parameters
x = final_sales_freq.keys()
y = final_sales_freq.values()
color = 'blue'
marker = 'o'
label = "Sales in euros"
linewidth = 2.5
markersize = 9.0
gridcolor = "grey"
xlabel = "Published year"
ylabel = "Average revenue \n of courses (x 1000)"
title = "Average sales of the courses per year"

# Draw the line graph
line_plot(x, y, color = color, marker = marker, label = label,
          linewidth = linewidth, markersize = markersize, xlabel = xlabel,
             ylabel = ylabel, title = title, xticks = years, gridcolor = gridcolor)
```

### Conclusion
The average amount of revenue per course has been falling over the last few years. Just like with the number of subscribers 

## Visulizations Head of Data Science
In this chapter, the visualization for the Head of Data Science wil be shown.

The visualizations will be as follows:
- Scatter plot of rating and price (`avg_rating`, `final_prize_amount_eur`)
- Scatter plot of subscribed students and price (`num_subscribers`, `final_prize_amount_eur`)
- Scatter plot of discount and rating (`discount_price__amount`, `avg_rating`)
- Scatter plot of discount and number of subscribed students (`discount_price__amount`, `num_subscribers`) 
- Average percentage of discounts per year (`published_year`, `discount_per`)

### Correlation between rating and price

```python
# Get the data
ratings = np.round(np.array(data['avg_rating']), 2)
price = np.array(np.round(data['final_prize_amount_eur'], 2))
```

```python
# Set the parameters
x = price
y = ratings
alpha = 0.15
color = "green"
title = "Scatter plot of price vs ratings"
xlabel = "Price (in EUR)"
ylabel = "Rating course"
label = "Rating"

# Set xticks to create better overview of the plot
xticks = [x for x in range(0, 131, 10)]

# Create the scatter plot
scatter_plot(x, y, color, alpha, title, xlabel, ylabel, label = label, xticks = xticks)
```

### Findings
Both cheap and expensive course provide high and low rated courses. Range of rating appears smaller between 40 and 60 euros.

### Correlation between subscribed students and price

```python
# Get the data
num_subscribers = np.array(data['num_subscribers'] / 1000)
price = np.array(np.round(data['final_prize_amount_eur']))
```

```python
x = price
y = num_subscribers
alpha = 0.1
color = "black"
title = "Scatter plot of price vs subscribed student"
xlabel = "Price (EUR)"
ylabel = "Amount of subscribed students (x 1000)"
label = "Subscribers"
scatter_plot(x, y, color, alpha, title, xlabel, ylabel, label)
```

### Findings
There seems to be no correlation between price and amount of student subscribed. Around 100 euros, there is a group of courses that have some outliers.

### Correlation between discount and rating

```python
# Get the data
dis_per = np.array(np.round(data['discount_per'], 2))
ratings = np.array(np.round(data['avg_rating'], 2))

# Set xticks to create better overview of the plot
xticks = [x for x in range(0, 61, 5)]

# Set the parameters
color = "red"
alpha = 0.1
title = "Discount vs rating"
xlabel = "Discount (%)"
ylabel = "Ratings"
label = "Rating"

# Create the scatter plot
scatter_plot(dis_per, ratings, color, alpha, title, xlabel, ylabel, label = label, xticks = xticks)
```

### Findings
There appears to be no correlation between discount and rating. There seems to be a relatively smaller range of rating around 55%

### Correlation between discount and number of subscribed students

```python
# Get the data
dis_per = np.array(np.round(data['discount_per'], 2))
num_subscribers = np.array(np.round(data['num_subscribers'] / 1000))

# Set xticks to create better overview of the plot
xticks = [x for x in range(0, 60, 5)]

# Set the parameters
color = "blue"
alpha = 0.2
title = "Discount vs number of subscribed students"
xlabel = "Discount (%)"
ylabel = "Number of subscribers (x 1000)"
label = "Subscribers"

# Create the scatter plot
scatter_plot(dis_per, num_subscribers, color, alpha, title, xlabel, ylabel, label = label, xticks = xticks)
```

### Findings
There seems to be no correlation at all between discount and number of subscribed students to a course. Widest range of subscribed students is at 5% discount.

### Average discount percentage over time

```python
# Get the data ready
dis_per = np.array(np.round(data['discount_per'], 2))
published_year = np.array(data['published_year'])

# Set the freq table
final_dis_freq = freq_table_avg(data, published_year, dis_per)
print(final_dis_freq)
```

```python
x = final_dis_freq.keys()
y = final_dis_freq.values()
color = "red"
marker = 'o'
label = "Discount (%)"
linewidth = 2.5
markersize = 9.0
xlabel = "Year"
ylabel = "Discount percentage"
title = "Average discount percentage over time"
xticks = years
gridcolor = 'green'

line_plot(x, y, color, marker, label, linewidth, markersize, xlabel,
             ylabel, title, xticks, gridcolor)
```

### Findings
The average discount has risen to above 22% at 2011. After that, it gradually fell until 2017. Since then, the average discount got gradually higher.

## Conclusion
Based on our visualization, we can conclude that there's no correlation between discount and the combination of rating and number of subscribed students. This is despite the fact that some course are offered at a discount of over more than 50%. We can also see that the quality of the courses have been stable, due to the solid ratings over the last decade.

My recommendation is to stop offering huge discount to courses and keep it at a maximum of 30%. This will generate more revenue per course and probably in total, since we couldn't find a correlation between discount and number of subscribed students.
